```python
import pandas as pd

df = pd.read_csv('dados_corrigidos/API_SE.PRM.UNER.ZS_DS2_en_csv_v2_4492680.csv', sep=',', encoding='latin1', skiprows=4, decimal='.')

df_br = df[df['Country Code'] == 'BRA']
df_al = df[df['Country Code'] == 'DEU']
df_gu = df[df['Country Code'] == 'GTM']

import matplotlib.pyplot as plt
plt.style.use('default')
import seaborn as sns

anos = range(2012,2020)
anos = [num for num in anos]

valores_p_ano_br = []
for ano in anos: #tirar o 10 em 10 para mais precisa
        valores_p_ano_br.append(df_br[str(ano)])

valores_p_ano_al = []
for ano in anos: #tirar o 10 em 10 para mais precisa
        valores_p_ano_al.append(df_al[str(ano)])

valores_p_ano_gu = []
for ano in anos: #tirar o 10 em 10 para mais precisa
        valores_p_ano_gu.append(df_gu[str(ano)])


plt.figure(figsize=(10,7))
plt.plot(anos,valores_p_ano_gu,label='Guatemala')
plt.plot(anos,valores_p_ano_al,label='Alemanha')
plt.plot(anos,valores_p_ano_br,label='Brasil')
plt.title('Porcentagem de Crianças Fora da Escola por País (Primário)')
plt.xlabel('Anos')
plt.ylabel('Quantidade de Crianças (%)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
import seaborn as sns

def trata_dados(br,al,gu):
    valores_p_ano_br = []
    for ano in anos: 
        valores_p_ano_br.append(br[str(ano)])

    valores_p_ano_al = []
    for ano in anos: 
        valores_p_ano_al.append(al[str(ano)])

    valores_p_ano_gu = []
    for ano in anos: 
        valores_p_ano_gu.append(gu[str(ano)])
    return valores_p_ano_br, valores_p_ano_al, valores_p_ano_gu
    
     
df = pd.read_csv('dados_APS2/API_SN.ITK.DEFC.ZS_DS2_en_csv_v2_4493905.csv', sep=',', encoding='latin1', skiprows=4, decimal='.')
df_br = df[df['Country Code'] == 'BRA']
df_al = df[df['Country Code'] == 'DEU']
df_gu = df[df['Country Code'] == 'GTM']

brasil,alemanha,guatemala = trata_dados(df_br, df_al, df_gu)

df = pd.read_csv('dados_corrigidos/API_SE.PRM.UNER.ZS_DS2_en_csv_v2_4492680.csv', sep=',', encoding='latin1', skiprows=4, decimal='.')
_df_br = df[df['Country Code'] == 'BRA']
_df_al = df[df['Country Code'] == 'DEU']
_df_gu = df[df['Country Code'] == 'GTM']

_brasil,_alemanha,_guatemala = trata_dados(_df_br, _df_al, _df_gu)

anos = range(2012,2020)
anos = [num for num in anos]

plt.figure(figsize=(15,10))                                                              
plt.plot(anos,alemanha,label='Desnutrição: Alemanha', color='red')
plt.plot(anos,_alemanha,label='Crianças fora da escola: Alemanha', color='red', linestyle='dashdot')

plt.plot(anos,_brasil,label='Crianças fora da escola: Brasil', color='dodgerblue', linestyle='dashdot')
plt.plot(anos,brasil,label='Desnutrição: Brasil', color='dodgerblue')

plt.plot(anos,guatemala,label='Desnutrição: Guatemala', color='green')
plt.plot(anos,_guatemala,label='Crianças fora da escola: Guatemala', color='green', linestyle='dashdot')

plt.title('Relação entre desnutrição e crianças fora da escola. Dados da Alemanha, Brasil e Guatemala.')
plt.xlabel('Anos')
plt.ylabel('Porcentagem (%)')
plt.legend()
plt.show()
```

```python
# Concatenado Brasil
df_br.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 66'], axis=1, inplace=True)
df_br.set_index('Country Name', inplace=True)
df_br = df_br.transpose()

_df_br.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 66'], axis=1, inplace=True)
_df_br.set_index('Country Name', inplace=True)
_df_br = _df_br.transpose()

brasil_concat = pd.concat([df_br, _df_br], axis=1)
brasil_concat.set_axis(['Desnutrição','Crianças fora da escola'], axis=1, inplace=True)
```
